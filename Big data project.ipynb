```python
#BPQ = Blood pressure questionnaire data 
bpq = pd.read_sas("BPQ_J.XPT")
bpq.columns
```

```python
#SMQ_J = Smoking cigarette questionnaire 
smq = pd.read_sas("SMQ_J.XPT")
smq.columns
```

```python
#Drop columns 
smq = smq.drop(columns = ['SMD030', 'SMQ050Q', 'SMQ050U', 'SMD057',
       'SMQ078', 'SMD641', 'SMD650', 'SMD093', 'SMDUPCA', 'SMD100BR',
       'SMD100FL', 'SMD100MN', 'SMD100LN', 'SMD100TR', 'SMD100NI', 'SMD100CO',
       'SMQ621', 'SMD630', 'SMQ661', 'SMQ665A', 'SMQ665B', 'SMQ665C',
       'SMQ665D', 'SMQ670', 'SMQ848', 'SMQ852Q', 'SMQ852U', 'SMQ895',
       'SMQ905', 'SMQ915', 'SMAQUEX2'])
smq.columns
```

```python
#Rename columns 
smq = smq.rename({'SEQN': 'ParticipantID',
                  'SMQ020': '100Cigs',
                  'SMQ040': 'Smoke_Cigs', #1 = Yes, 2 = No
                  'SMQ890': 'Smoke_Cigar',
                  'SMQ900': 'E_cig',
                  'SMQ910': 'Smokeless_tobacco'},axis = 1)
smq.head()
```

```python
#Diabetes data 
diab = pd.read_sas("DIQ_J.XPT")
diab.columns
```

```python
#Drop columns
diab = diab.drop(columns = ['DIQ172', 'DIQ175C', 'DIQ175D', 'DIQ175E', 'DIQ175F', 'DIQ175G',
       'DIQ175H', 'DIQ175I', 'DIQ175J', 'DIQ175K', 'DIQ175L', 'DIQ175M',
       'DIQ175N', 'DIQ175O', 'DIQ175P', 'DIQ175Q', 'DIQ175R', 'DIQ175S',
       'DIQ175T', 'DIQ175U', 'DIQ175V', 'DIQ175W', 'DIQ175X',
       'DIQ050', 'DID060', 'DIQ060U', 'DIQ070', 'DIQ230', 'DIQ240', 'DID250',
       'DID260', 'DIQ260U', 'DIQ275', 'DIQ280', 'DIQ291', 'DIQ300S', 'DIQ300D',
       'DID310S', 'DID310D', 'DID320', 'DID330', 'DID341', 'DID350', 'DIQ350U',
       'DIQ360', 'DIQ080', 'DIQ175B'])
```

```python
#Rename columns 
diab = diab.rename({'SEQN': 'ParticipantID',
                    'DIQ010': 'Diagnosis', #label 
                    'DID040': 'Diagnosed_age',
                    'DIQ160': 'Prediabetes',
                    'DIQ170': 'Diabetes_risk',
                    'DIQ175A': 'Fam_hist',
                    'DIQ180': 'Blood_test' #1 = yes, 2 = no
}, axis = 1)
diab.head()
```

```python
#Combine all dataframes
df = demo.set_index('ParticipantID').join(bp.set_index('ParticipantID')).join(diet_1.set_index('ParticipantID')).join(bm.set_index('ParticipantID')).join(ins.set_index('ParticipantID')).join(glu.set_index('ParticipantID')).join(alc.set_index('ParticipantID')).join(bpq.set_index('ParticipantID')).join(pa.set_index('ParticipantID')).join(smq.set_index('ParticipantID')).join(diab.set_index('ParticipantID'))
```
